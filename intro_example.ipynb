### Plot a white noise process

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e = np.random.randn(100) #call randn from subpackage random
plt.plot(e)
plt.show()
```

#### Alternate Version 1: For Loop

```python
ts_n = 100
e=[]

for i in range(ts_n):
    e.append(np.random.randn())

plt.plot(e)
plt.title('Alternative Version 1: White Noise Process')
plt.show()
```

### While Loops

Iteration based on condition

```python
ts_n = 100
e=[]
counter = 0

while counter < ts_n:
    e.append(np.random.randn())
    counter +=1

plt.plot(e)
plt.title('Alternative Version 2: White Noise Process')
plt.show()
```

### Plot bank account balances over time

Given the following assumptions:

    (1) There are no withdraws over the time period, the last date of which is denoted by T.

    (2) The initial balance is b0 and the interest rate is r.

    (3) The balance updates from period t to t+1 according to bt+1=(1+r)bt.


Store results in an array and plot the balances

```python
r = 0.025 # interest rate
T = 50 # end date
b = np.empty(T+1) # instantiate array
b[0] = 10 # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.title('Bank Account Balance Over Time')
plt.legend()
plt.show()
```
